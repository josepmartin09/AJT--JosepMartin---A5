```python
class song:

    def __init__ ( self, titol, autor, durada):
        self.titol = titol
        self.autor = autor
        self.durada = durada

    def info ( self ):
        return f"{self.titol} - {self.autor} {self.durada} segons"
    

class playlist:

    def __init__ ( self, song_list):
        self.song_list = song_list  
        song_list = []

    def afegir_canco ( self, canco):
        self.song_list.append(canco)

    def durada_total ( self):
        total = 0
        for i in self.song_list:
            total += i.durada
        return total

    def mostrar_playlist ( self):
        for n in self.song_list:
            print ( n.info())

    def canco_mes_llarga ( self):
        canco_mes_llarga = self.song_list[0]
        for x in self.song_list:
            if x.durada > canco_mes_llarga.durada:
                canco_mes_llarga = x
        return canco_mes_llarga.info()


canco1 = song ( "Nothing but a G Thang", "Dr. Dre", 267)

canco2 = song ( "All of the Lights", "Kanye West", 300)

playlist1 = playlist ( [canco1, canco2])

canco3 = song ( "Empire State of Mind", "Jay-Z", 261.6)  

playlist1.afegir_canco ( canco3)

print ( playlist1.durada_total())
print ( playlist1.canco_mes_llarga())
playlist1.mostrar_playlist()


```
